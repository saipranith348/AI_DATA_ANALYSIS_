```python
import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Customize plot
plt.figure(figsize=(8, 6))                        # Set figure size
plt.bar(products, sales, color='teal')            # Change bar color
plt.title('Product Sales', fontsize=16, fontweight='bold')  # Custom title font
plt.xlabel('Product', fontsize=12, fontstyle='italic')      # Custom X-axis label
plt.ylabel('Sales (in thousands)', fontsize=12, fontstyle='italic')  # Custom Y-axis label
plt.grid(True, linestyle='--', alpha=0.6)         # Add grid with style

# Show plot
plt.tight_layout()
plt.show()
```
