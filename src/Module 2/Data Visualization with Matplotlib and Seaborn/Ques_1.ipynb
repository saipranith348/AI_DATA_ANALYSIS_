```python
import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Input validation
if not products or not sales:
    raise ValueError("Product and sales lists must not be empty.")
if len(products) != len(sales):
    raise ValueError("Product and sales lists must be of the same length.")

# Create bar plot
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color='skyblue')
plt.title('Sales of Products (in thousands)')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
