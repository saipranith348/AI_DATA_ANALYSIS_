```python
import matplotlib.pyplot as plt

# Data
marks = [23, 45, 56, 78, 89, 90, 67, 88, 76, 59]

# Histogram
plt.hist(marks, bins=5, edgecolor='black')  # You can adjust 'bins' for more/less granularity

# Labels and title
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Student Marks")

# Show plot
plt.show()
```
