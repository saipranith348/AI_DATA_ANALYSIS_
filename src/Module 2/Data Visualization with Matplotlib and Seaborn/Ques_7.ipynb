```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]
sizes = [10, 15, 13, 18, 12]

# Create DataFrame
data = pd.DataFrame({
    'Height': heights,
    'Weight': weights,
    'Size': sizes
})

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Height', y='Weight', size='Size', sizes=(50, 300), legend=False, color='purple')
plt.title('Height vs. Weight Scatter Plot (Size Encoded)', fontsize=14)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
